```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
plt.figure(figsize = (20,13))
from sklearn.tree import plot_tree
plot_tree(clf)
```

## If a decision tree is long enough then it might be the case of overfitting


```python
from sklearn.tree import DecisionTreeClassifier
def analyzer(max_depth):
    df2 = pd.read_csv('train_data.csv')
    X = df2.iloc[:,2:4].values
    y = df2.iloc[:,-1].values
    cla = DecisionTreeClassifier(max_depth = max_depth)
    cla.fit(X,y)
    
    a = np.arange(start = X[:,0].min() - 1,stop = X[:,0].max() + 1,step = 0.5)
    b = np.arange(start = X[:,1].min() - 1,stop = X[:,1].max() + 1,step = 0.5)
    #print(a.shape,b.shape)
    XX,YY = np.meshgrid(a,b)
    #print(XX.shape)
    input_array = np.array([XX.ravel(),YY.ravel()]).T
    labels = cla.predict(input_array)
    #print(labels.shape)
    plt.contourf(XX,YY,labels.reshape(XX.shape))
    plt.scatter(X[:,0],X[:,1],c = y)
    plt.show()
```

```python
analyzer(200) ##overfitting
```

```python
analyzer(3) #Underfitting
```
